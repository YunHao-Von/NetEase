```python
from data_get import *
import pandas as pd
import numpy as np
import multiprocessing as mp
```

```python
user_id_list = pd.read_csv('data_from_netease/users.csv')['user_id'].tolist()
user_id_list
```

```python
def process_fit(user_id):
    dataspider = User_spider()
    user_id_list=dataspider.usersongs_spider_fit(user_id)
    user_id_list = np.array(user_id_list)
    np.save("data_from_netease/data_files/"+str(user_id)+".npy",user_id_list)
```

```python
mp.cpu_count()
print(len(user_id_list))
import os
files_list = os.listdir('data_from_netease/data_files')
success = list()
for file_name in files_list:
    success.append(int(file_name[:-4]))
ret_list = [item for item in user_id_list if item not in success]
user_id_list = ret_list
print(len(user_id_list))
```

```python
pool = mp.Pool()
pool.map(process_fit,user_id_list)
pool.close()
pool.join()
```

```python
count = 0
for id in success:
    if len(np.load("data_from_netease/data_files/"+str(id)+".npy").tolist()) == 0:
        os.remove("data_from_netease/data_files/"+str(id)+".npy")
        count += 1
count
```
