```python
import os
import numpy as np
files_list = os.listdir('data_from_netease/data_files')
success = list()
for file_name in files_list:
    success.append(int(file_name[:-4]))
```

```python
adjacency_matrix = np.zeros((len(success),len(success)))
# adjacency_matrix[0][0] =1 
adjacency_matrix
```

```python
for i in range(len(success)):
    a = np.load("data_from_netease/data_files/"+str(success[i])+".npy").tolist()
    for j in range(len(success)):
        b = np.load("data_from_netease/data_files/"+str(success[j])+".npy").tolist()
        length_inner = list(set(a).intersection(set(b)))
        adjacency_matrix[i][j]=len(length_inner)/len(a)
```

```python
np.save("adjacency_matrix.npy",adjacency_matrix)
```

```python
np.count_nonzero(adjacency_matrix)
```

```python
draw_array = adjacency_matrix[:20,:20]
np.count_nonzero(draw_array)

from pyecharts import options as opts
from pyecharts.charts import Graph
nodes_data = list()
links_data = list()
for i in range(20):
    nodes_data.append(opts.GraphNode(name="节点"+str(i),symbol_size=40))
    for j in range(20):
        links_data.append(opts.GraphLink(source="节点"+str(i), target="节点"+str(j), value=draw_array[i][j]))
        print(draw_array[i][j])
c = (
    Graph()
    .add(
        "",
        nodes_data,
        links_data,
        repulsion=40000,
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="网易云音乐用户谱系图（局部）")
    )
    .render("users_connection.html")
)
```
